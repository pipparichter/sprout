```python
import json 
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

%load_ext autoreload 
%autoreload 2
```

```python


def load_results(path:str='../results/dataset_train-mlp_x10_0.results.json'):
    with open(path, 'r') as f:
        results = json.load(f)

    ids = results['ids'].copy()
    del results['ids']

    results_df = list()
    for model_id, results_ in results.items():
        df = pd.DataFrame(results_).assign(model_id=model_id)
        df['id'] = ids
        results_df.append(df)
    results_df = pd.concat(results_df)
    return results_df

load_results()
```
